```python
berri_bikes = data[["Berri1"]].copy()
berri_bikes[:5]
```

```python
berri_bikes.index
```

```python
berri_bikes.index.weekday # Número del día de la semana 0 - 6
```

```python
berri_bikes["weekday"] = berri_bikes.index.weekday
berri_bikes[:5]
```

```python
berri_bikes.loc[:, "weekday"]
```

```python
grupos = berri_bikes.groupby("weekday")
grupos.groups
```

```python
cuentas = grupos.aggregate(sum)
cuentas
```

```python
cuentas.index = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
# Reescribe los valores de index con el siguiente array
cuentas
```

```python
cuentas.plot(kind="bar")
```
