```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from joblib import load

in_file = 'data_clean/MFC_ActualizacionJul2023_clean.pkl'
with open(in_file, 'rb') as f:
    df = load(f)
    
df.tail()
```

```python
pd.pivot_table(df, index='Restos_tipo', columns='Sexo', values='Edad_int', aggfunc='mean')
```

```python
df['Fecha_inhumación_date_año'] = df['Fecha_inhumación_date'].dt.year
df['Fecha_defunción_date_año'] = df['Fecha_defunción_date'].dt.year
```

```python
pd.pivot_table(df, index='Restos_tipo', columns='Fecha_inhumación_date_año',
              aggfunc='size', dropna=False
              )
```

```python
pd.pivot_table(df, index=['Sexo', 'Restos_tipo'],
              columns=['Fecha_inhumación_date_año'],
              aggfunc='size'
              )
```

```python
pd.pivot_table(df, values='Tipo_institución_origen',
              index='Restos_tipo', columns='Fecha_inhumación_date_año',
              aggfunc=pd.Series.mode, dropna=False
              )
```

```python
def unir_textos_unicos_en_serie(series):
    textos = series.sort_values().unique()
    textos = ', '.join(textos)
    return textos

pd.pivot_table(df, values='Fecha_inhumación_date_diasemana',
              index='Restos_tipo', columns='Fecha_inhumación_date_año',
              aggfunc=unir_textos_unicos_en_serie,
              dropna=False
              )
```

```python
df_edad_promedio = pd.pivot_table(df, values='Edad_int',
                                 index='Restos_tipo', columns='Sexo'
                                 )
df_edad_promedio
```

```python
df_edad_promedio = df_edad_promedio.reset_index()
df_edad_promedio
```

```python
df_edad_promedio = pd.melt(df_edad_promedio, value_name='Edad_promedio',
                          id_vars='Restos_tipo', value_vars=['Femenino', 'Indeterminado', 'Masculino'],
                          )
df_edad_promedio
```

```python
df[['Restos_tipo', 'Edad_int']].groupby(by='Restos_tipo')['Edad_int'] \
    .agg(['min', 'max', pd.Series.mode, 'mean', 'std']) \
    .rename(columns = {
        'min': 'edad mínima', 'max': 'edad máxima',
        'mode': 'moda edad', 'mean': 'promedio edad',
        'std': 'desviación estandar',
    })
```
