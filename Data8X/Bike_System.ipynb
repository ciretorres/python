# Bike system

```python
trips = Table.read_table('db/2017-fordgobike-tripdata.csv')
trips
```

```python
commute = trips.where('duration_sec', are.below(1800))
commute.hist('duration_sec')
```

```python
commute.hist('duration_sec', bins=60, unit='second')
```

```python
(550 - 250) * 0.14
```

```python
commute.hist('duration_sec', bins=np.arange(1801), unit='second')
```

```python
starts = commute.group('start_station_name').sort('count', descending=True)
starts
```

```python
commute.pivot('start_station_name', 'end_station_name')
```

```python
duration = trips.select(4, 8, 0)
duration
```

```python
shortest = duration.group([0, 1], min)
shortest
```

```python
from_cc = shortest.where(0, are.containing('45th St at MLK Jr Way')).sort(2)
from_cc
```

```python
stations = Table.read_table('db/station.csv')
stations
```

```python
Marker.map_table(stations.select('lat', 'long', 'name'))
```

```python
sf = stations.where('landmark', 'San Francisco')
Circle.map_table(sf.select('lat', 'long', 'name'), color='green', radius=200)
```

```python
from_cc
```

```python
colors = stations.group('landmark').with_column(
    'color', make_array('blue', 'red', 'green', 'orange', 'purple')
)
colors
```

```python
colored = stations.join('landmark', colors).select('lat', 'long', 'name', 'color')
Marker.map_table(colored)
```

```python
station_starts = stations.join('name', starts, 'start_station_name')
station_starts
```

```python
circle = station_starts.select('lat', 'long', 'name').with_columns(
    'color', 'blue',
    'area', station_starts.column('count') * 1000
)
circle
```

```python
Circle.map_table(circle)
```
