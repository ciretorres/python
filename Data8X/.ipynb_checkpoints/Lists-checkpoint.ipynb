## Two-dimensional array

```python
arr = []
arr.append([])
arr[0].append('aa1')
arr[0].append('aa2')
arr[0].append('aa3')
arr.append([])
arr[1].append('bb1')
arr[1].append('bb2')
arr[1].append('bb3')
arr[1][0]
```

```python
globe = []

for i in np.arange(0, 50):    
    globe.append([])
    for j in np.arange(0, 50):    
        globe[i].append('aa' + str(j)) 
        #print('globe[' + str(i) + '][' + str(j) + '] = ' + globe[i][j])
    
#globe[2][5] = 'hola'
```

```python
arr = []

for i in np.arange(0, 50):    
    arr.append([])
    for j in np.arange(0, 50):    
        arr[i].append('aa' + str(j)) 
        print('arr[' + str(i) + '][' + str(j) + '] = ' + arr[i][j])
    
#arr[2][5]
```
